## Next Step: Whittle Down Based on Median Value in The Past Year
- Here we grab zipcodes with a range of between 35% and 65% (15% above and below the median value from the past 12 months)

```python
# Grab the mean value of the last 12 months

bottom25zips['yr_avg']=bottom25zips.iloc[:,-12:].mean(skipna=True, axis=1)

#Get zipcodes with an average value 15% above the median and 15% below.
print(bottom25zips['yr_avg'].describe(),'\n')
#Calculate the 60% cutoff value (15% above).
q_65 = bottom25zips['yr_avg'].quantile(q=0.65)
print(f'Average Value 15% above median: {round(q_65,2)}')
#Calculate the 35% cutoff value (15% below).
q_35 = bottom25zips['yr_avg'].quantile(q=0.35)
print(f'Average Value 15% below median: {round(q_35,2)}')
#Get data frame with selected zipcodes.
zips_we_want = bottom25zips[(bottom25zips['yr_avg']<q_65) & (bottom25zips['yr_avg']>q_35)]
print(f'Amount of zipcodes: {len(zips_we_want)}')
```

### Next, we calculate the following values for our ZipCodes:
- ROI: Profit/Investment
- Standard Deviation of Monthly Values
- Historical Mean Value
- The Coefficient of Variation (Measure of Relative Variability.)
    - This is the the ratio of the standard deviation to the mean (average).

```python
#Calculate ROI
zips_we_want['ROI']= (zips_we_want['2018-04']/zips_we_want['1996-04'])-1
#Calculate standard deviation of monthly values
zips_we_want['std']=zips_we_want.loc[:,'1996-04':'2018-04'].std(skipna=True, axis=1)
#Calculate historical mean value
zips_we_want['mean']=zips_we_want.loc[:,'1996-04':'2018-04'].mean(skipna=True, axis=1)
#Calculate coefficient of variance
zips_we_want['CV']=zips_we_want['std']/zips_we_want['mean']
#Show calculated values
zips_we_want[['RegionName','std','mean','ROI','CV']].head()
```

```python
#Descriptive statistics of coefficients of variance.
print(zips_we_want.CV.describe())
#Define upper limit of CV: .6 
upper_cv = zips_we_want.CV.quantile(.6)
print(f'\nCV upper limit: {upper_cv}')
#Get the 5 zipcodes with highest ROIs within our risk profile.
top5zips = zips_we_want[zips_we_want['CV']<upper_cv].sort_values('ROI',axis=0,ascending=False)[:5]
print('\n Best 5 Zipcodes:')
top5zips[['RegionName','ROI','CV']]
```

```python
#Get Location Names
best5_zipcodes = list(top5zips.RegionName.values)
for i in best5_zipcodes:
    city = df[df['RegionName']==i].City.values[0]
    state = df[df['RegionName']==i].State.values[0]
    print(f'Zipcode : {i} \nLocation: {city}, {state}\n')
```

# Helper Functions

```python
# Obtain dates from our data:
def get_datetimes(df):
    return pd.to_datetime(df.columns.values[1:], format='%Y-%m')
```

```python
#Convert the data into long format.
def melt_data(df):
    melted = pd.melt(df, id_vars=['RegionName'], var_name='time')
    melted['time'] = pd.to_datetime(melted['time'], infer_datetime_format=True)
    melted = melted.dropna(subset=['value'])
    return melted
```

```python
TS_zc5 = top5zips.drop(['yr_avg','std','mean','ROI','CV'],axis=1)
TS_zc5 = melt_data(TS_zc5).set_index('time')
print('Time series data for the 5 zipcodes:\n',TS_zc5.head())
#Create individualized time series for each zipcode.
#List containing the 5 different time series.
df_ts = []
for zipcode in TS_zc5.RegionName.unique():
    #Create separate dataframes for each zipcode with a monthly frequency.
    df = TS_zc5[TS_zc5['RegionName']==zipcode].asfreq('MS')
    df_ts.append(df)
print('\nZipcode 48894 time series:')
df_ts[0].head()
```

```python
for i in range(len(df_ts)):
    print(f'Value descriptive statistics for zipcode {df_ts[i].RegionName[0]}:')
    print(f'{df_ts[i].value.describe()}\n')
```

```python
for i in range(5):
    df_ts[i].value.plot(label=df_ts[i].RegionName[0],figsize=(15,6))
    plt.legend()
```

```python
df_ts[0].head()
```

```python
#Calculate monthly returns in new column 'returns' for each zipcode.
for zipcode in range(len(df_ts)):
    df_ts[zipcode]['returns']=np.nan*len(df_ts[zipcode])
    for i in range(len(df_ts[zipcode])-1):
        df_ts[zipcode]['returns'][i+1]= (df_ts[zipcode].value.iloc[i+1] / df_ts[zipcode].value.iloc[i]) - 1
```

```python
df_ts[0].head()
```

```python
#Plot the monthly returns of each zipcode
for i in range(len(df_ts)):
    df_ts[i].returns.plot(figsize=(11,5), color = 'b')
    plt.title(f'Zipcode: {df_ts[i].RegionName[0]}')
    plt.xlabel('Date')
    plt.ylabel('Returns (%)')
    plt.legend(loc='best')
    plt.show()
```

```python
#Plot each of the zipcodes' returns with their respective rolling mean and rolling standard deviation.
#Visually test for stationarity.
for i in range(len(df_ts)):
    rollingmean = df_ts[i].returns.rolling(window = 12, center = False).mean()
    rollingstd = df_ts[i].returns.rolling(window = 12, center = False).std()
    fig = plt.figure(figsize=(11,5))
    orig = plt.plot(df_ts[i].returns, color='blue',label='Original')
    mean = plt.plot(rollingmean, color='red', label='Rolling Mean')
    std = plt.plot(rollingstd, color='black', label = 'Rolling Std')
    plt.xlabel('Dates')
    plt.ylabel('Returns (%)')
    plt.legend(loc='best')
    plt.title(f'Rolling Mean & Standard Deviation for Zipcode: {df_ts[i].RegionName[0]}')
    plt.show()
```

A few of these, 48894 and 40008 in particular look stationary--56360 and 49339 look like they have seasonality and I'm not sure about 27019--but will need further investigation with a Dickey-Fuller Test

```python
for i in range(5):
    results = adfuller(df_ts[i].returns.dropna())
    print(f'ADFuller test p-value for zipcode: {df_ts[i].RegionName[0]}')
    print('p-value:',results[1])
    if results[1]>0.05:
        print('Fail to reject the null hypothesis. Data is not stationary.\n')
    else:
        print('Reject the null hypothesis. Data is stationary.\n')
```

Let's test to see what will happen if we use differencing on the non-stationary zip codes!

```python
# Use Differencing on the two zipcodes' returns that are not stationary:
# .diff () is used to find the first discrete difference of objects over a given axis. 
for i in [1,3]:
    #Perform adfuller test and drop NaN values created when calculating monthly returns.
    results = adfuller(df_ts[i].returns.diff().dropna())
    print(f'ADFuller test p-value for zipcode: {df_ts[i].RegionName[0]}')
    print('p-value:',results[1])
    if results[1]>0.05:
        print('Fail to reject the null hypothesis. Data is not stationary.\n')
    else:
        print('Reject the null hypothesis. Data is stationary.\n')
```

```python
# Create function to display ACF and PACF. Lags=60 (5 years)
def acf_pacf(df,alags=60,plags=60):
    #Create figure
    fig,(ax1,ax2) = plt.subplots(2,1,figsize=(13,8))
    #Make ACF plot
    plot_acf(df,lags=alags, zero=False,ax=ax1)
    #Make PACF plot
    plot_pacf(df,lags=plags, ax=ax2)
    plt.show()
```

```python
def train_test(df):
    #Set training data before 2015
    train = df[:'2015-01']
    #Set test data starting 2016
    test = df['2015-01':]
    return train, test
```

```python
def fit_model(df,pdq=(1,0,1),pdqs=(0,0,0,1)):
    train, test = train_test(df)
    model = SARIMAX(train,order=pdq,seasonal_order=pdqs)
    results = model.fit()
    results.summary
    residuals = results.resid
    print(results.summary())
    results.plot_diagnostics(figsize=(11,8))
    plt.show();
    return train, test, results
```

```python
#Fit the SARIMA model and get results.
pdq = (1,0,1)
pdqs = (0,0,2,12)
train, test, results = fit_model(TS_48894,pdq=pdq,pdqs=pdqs)
```

```python
def test_RMSE(df,pdq,pdqs, display=True):
    X = df.values
    train, test = X[:-36],X[-36:]
    history = [x for x in train]
    predictions = []
    for t in range(len(test)):
        model = SARIMAX(history, order=pdq,seasonal_order=pdqs)
        fit_model = model.fit(disp=0)
        output = fit_model.forecast()
        yhat = output[0]
        predictions.append(yhat)
        history.append(test[t])
    rmse = sqrt(MSE(test, predictions))
    print('SARIMA model RMSE on test data: %.5f' % rmse)
    if display:
        plt.figure(figsize=(13,6))
        plt.title('Actual Test Data vs. Predictions')
        plt.plot(history[-36:],label='Actual', color='b')
        plt.plot(predictions,label='Predictions',color='r')
        plt.legend(loc='best')
        plt.show()

def train_RMSE(train, results, display = True):
    train_pred = results.predict(-36)
    rmse = sqrt(MSE(train[-36:],train_pred))
    print(f'SARIMA model RMSE on train data: %.5f' % rmse)
    if display:
        plt.figure(figsize=(13,6))
        train[-60:].plot(label='Actual',color='b')
        train_pred.plot(label='Predicted',color='r')
        plt.legend(loc='best')
        plt.title('Actual Train Data vs. Predicted Returns')
        plt.show()
```

```python
def forecast_model(df,pdq,pdqs, display=True,zc='input zipcode'):
    model = SARIMAX(df, order=pdq,seasonal_order=pdqs)
    fit_model = model.fit()
    output = fit_model.get_prediction(start='2018-04',end='2028-04', dynamic=True)
    forecast_ci = output.conf_int()
    if display:
        fig, ax = plt.subplots(figsize=(13,6))
        output.predicted_mean.plot(label='Forecast')
        ax.fill_between(forecast_ci.index,forecast_ci.iloc[:, 0],forecast_ci.iloc[:, 1],
                        color='k', alpha=.25,label='Conf Interval')
        plt.title('Forecast of Monthly Returns')
        plt.xlabel('Time')
        plt.legend(loc='best')
        plt.show()
    year_1= (1+output.predicted_mean[:12]).prod()-1
    year_3=(1+output.predicted_mean[:36]).prod()-1
    year_5= (1+output.predicted_mean[:60]).prod()-1
    year_10=(1+output.predicted_mean).prod()-1
    print(f'Total expected return in 1 year: {round(year_1*100,2)}%')
    print(f'Total expected return in 3 years: {round(year_3*100,2)}%')
    print(f'Total expected return in 5 year: {round(year_5*100,2)}%')
    print(f'Total expected return in 10 years: {round(year_10*100,2)}%')
    tot_ret = [zc,year_1,year_3,year_5,year_10]
    return tot_ret
```

```python
ret_48894=forecast_model(TS_48894,pdq=pdq,pdqs=pdqs,zc=48894)
```
